```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt

def func(x,m,b):
    return m * x + b

q = np.array([7.96,9.65,11.38,12.74,20.81])*1e-19
n = np.array([5,6,7,8,13])
p, pcov = opt.curve_fit(func, n, q, p0=[0,0])
perr = np.sqrt(np.diag(pcov))
x = np.linspace(0,13,100)

plt.figure(dpi=400)
plt.grid()
plt.errorbar(n,q,marker = '+',ls='',lw=0.5)
plt.plot(x,func(x,p[0],p[1]),lw = 1)

print(p[0],perr[0])
```
