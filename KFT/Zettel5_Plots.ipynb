```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mco

a = 1
q = 1

x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
rx, ry = np.meshgrid(x,y)
q1 = q
q2 = -q
q3 = q
q4 = -q

def potential(a,x,y):
    return (1/(np.sqrt((x - a/2)**2 + (y - a/2)**2)) + 1/(np.sqrt((x + a/2)**2 + (y + a/2)**2)) - 1/(np.sqrt((x-a/2)**2 + (y + a/2)**2)))- 1/(np.sqrt((x + a/2)**2 + (y-a/2)**2))

##berechnen des arrays
pot = q * potential(a,rx,ry)

##level
levels = np.linspace(-10,10,200)

cmap = mco.ListedColormap ( np.random.rand ( 256,3))

plt.figure(dpi=200)
plt.title("Äquipotentiallinien in xy-Ebene")
plt.ylabel("y")
plt.xlabel("x")
plt.axes().set_aspect('equal')
plt.contour(rx,ry,pot,levels = levels, cmap = 'twilight')
plt.savefig("HA5_1d.png")
```
