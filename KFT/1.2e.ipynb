```python
import numpy as np
import matplotlib.pyplot as plt


##strecke x
x = np.linspace(-0.1,0.1,100)

##Konstanten
c = 1
l = 1

##erste randbed
def f1(x,t):
    return np.exp(-(x+c*t)**2/(2*l**2))-np.exp(-(c*t)**2/(2*l**2))


##zweite
def f2(x,t):
    return np.exp(-(x+c*t)**2/(2*l**2))+x*c*t/(l**2)*np.exp(-(c*t)**2/(2*l**2))

plt.figure(dpi = 200)
plt.grid()
plt.title("Erste Randbedingung")
plt.xlabel("x")
plt.ylabel("$f_1(x,t)$")
plt.plot(x,f1(x,-0.1),linewidth = 1, label = "$t$=-0.1s")
plt.plot(x,f1(x,0),linewidth = 1, label = "$t$=0s")
plt.plot(x,f1(x,0.1),linewidth = 1,label = "$t$=0.1s")
plt.legend(loc='lower center')

plt.figure(dpi = 200)
plt.grid()
plt.title("Zweite Randbedingung")
plt.xlabel("x")
plt.ylabel("$f_2(x,t)$")
plt.plot(x,f2(x,-0.1),linewidth = 1, label = "$t$=-0.1s")
plt.plot(x,f2(x,0),linewidth = 1, label = "$t$=0s")
plt.plot(x,f2(x,0.05),linewidth = 1,label = "$t$=0.05s")
plt.legend(loc='lower center')
```
