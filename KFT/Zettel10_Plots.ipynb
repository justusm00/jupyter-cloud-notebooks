```python
import numpy as np
import matplotlib.pyplot as plt

c = 1
q = 1
m = 1
P0 = 0.8*c/m
E = 1
L = 1

x = np.linspace(0,L,1000)

def cosh(x):
    return 0.5*(np.exp(x)+np.exp(-x))

def z_rel(x):
    return c/(q*E)*np.sqrt(m**2*c**2+P0**2)*(cosh(x*q*E/(c*P0))-1)

def z(x):
    return 0.5*q*E*m**2/P0**2*x**2

plt.figure(dpi=400)
plt.title("nicht relativistisch")
plt.grid()
plt.xlabel("x [m]")
plt.ylabel("z [m]")
plt.plot(x,z(x),lw=1)
plt.savefig("nonrel.png")


x = np.linspace(0,L,1000)

plt.figure(dpi=400)
plt.title("relativistisch")
plt.grid()
plt.xlabel("x [m]")
plt.ylabel("z [m]")
plt.plot(x,z_rel(x),lw=1)
plt.savefig("rel.png")


```
