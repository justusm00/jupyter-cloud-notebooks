# Python Script zur Berechnung und graphischen Darstellung der Coulombkraft

Graphische Darstellung der Abstandsabhängigkeit beim Coulombgesetz. Wir betrachten eine Probeladung $e$ und eine Punktladung von 1 C. Wir nutzen *matplotlib*, um unsere Daten graphisch darzustellen:

```python
import matplotlib.pyplot as plt
```

Benötigte Konstanten sind $\pi$, $\epsilon_0$ und die Elementarladung. Wir können diese per Hand eingeben, oder einfach importieren:

```python
from scipy.constants import pi, epsilon_0, elementary_charge
```

Als Beweis, dass das funktioniert hat:

```python
print("pi =",pi)
print("epsilon_0 =",epsilon_0)
print("elementary_charge =",elementary_charge)
```

Der Proportionalitätsfaktor $f$ mit der Elementarladung und der Testladung von 1C ist:

```python
f = 1/(4*pi*epsilon_0) *elementary_charge*1
```

Wir können nun die Coulomb-Kraft als Funktion von $r$ berechnen. Dazu brauchen wir die Punkte an denen wir das machen wollen. Wir sparen den Punkt $r=0$ aus:

```python
r = [0.01*x for x in range(1,101)]
```

Was passiert hier?

*x in range(1,101)* nimmt nach und nach eine Zahl *x* aus dem Interval [1;101). Also 1,2 ... 100. Das nehmen wir mit 0.01 mal (*0.01 * x*) und packen das Ganze in eine praktische Liste mit [ ... ], die wir *r* nennen. Sie hat also die Punkte:

```python
print("Abstände r [m] =",r)
```

Auf die gleiche, geschickte Art und Weise können wir die Coulomb-Kraft berechnen:

```python
F_coulomb = [f/x**2 for x in r]
```

Wobei wir hier die für Python typische Kurzschreibweise *for x in r* verwendet haben, die bedeutet: nehme jedes Element *x* aus *r*.

Entsprechend das Ergebnis:

```python
print("Kraft F_coulomb [N] =",F_coulomb)
```

Zum Plotten: wir setzten die Größe der Graphik auf z.B. 12x6, also eine Querformat Darstellung:

```python
plt.figure(figsize=(12.,6.))

#linear plot
plt.subplot(121)
plt.plot(r,F_coulomb)
plt.ylabel('Kraft [N]')
plt.xlabel('Abstand [m]')
plt.yscale('linear')

#logarithmic plot
plt.subplot(122)
plt.plot(r,F_coulomb)
plt.ylabel('Kraft [N]')
plt.xlabel('Abstand [m]')
plt.yscale('log')
```

Man Beachte die Notation bei *subplot*. Die dreistellige Zahl sind eigentlich drei einstellige Zahlen, als0 1,2,1. Wobei die erste Stelle die Anzahl der Zeilen, die Zweite die Anzahl der Reihen und die Dritte die gerade benutzte Nummer des Graphen angibt.
