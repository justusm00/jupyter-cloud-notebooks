```python
import numpy as np
import matplotlib.pyplot as plt

##constants
C = 100e-9
L = 1e-3

#frequency
w = np.arange(0.1,1e6,1)



def rehsohnantzfrekwänz(w,R):
    a = w*L - 1/(w*C)
    return R/np.sqrt(R**2+a**2)


plt.figure(figsize = (8,6))
plt.grid()
plt.title("$U_a/U_e$ plottet against the frequency ")

i = 20
while(i < 200):
    plt.plot(w,rehsohnantzfrekwänz(w,i),linewidth = 1,label = "$R$ = %d$\,\Omega$"%i)
    i += 40
    
plt.legend(loc = "upper right")
```
