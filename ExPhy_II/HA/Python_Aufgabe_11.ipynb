**Hilfestellung**: Beachten Sie, dass $\sin (0) = 0$, das heißt hier würde durch $0$ geteilt. `numpy` wirft hier eine Warnung, aber fährt fort. Im Plot erscheinen diese Werte nicht. Die Abstrahlcharakteristik für $\theta=0$ ist 0. Um die Warnungen zu umgehen, den `numpy.linspace` so wählen, dass kein $\theta = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt

def E(theta,n):   ### n = vielfaches der Wellenlänge
    return abs(np.cos(n*np.pi*np.cos(theta))-np.cos(n*np.pi))/(2*np.pi*np.sin(abs(theta)))


theta = np.linspace(-np.pi,np.pi,98)
fig, ax = plt.subplots(3,2, figsize=(12,12),subplot_kw=dict(polar=True))
ax[0,0].plot(theta,E(theta,1/4))
ax[0,1].plot(theta,E(theta,1/2))
ax[1,0].plot(theta,E(theta,1))
ax[1,1].plot(theta,E(theta,3/2))
ax[2,0].plot(theta,E(theta,2))
plt.tight_layout()
```
