```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,12*np.pi,1000)

def spiral(t):
    real = np.zeros(len(t))
    im = np. zeros(len(t))
    for i in range(len(t)):
        if(t[i] <= 6 *np.pi):
            s = t[i] * np.exp(1j * t[i])
            real[i] = s.real
            im[i] = s.imag
        else: real[i] = 6 *np.pi - (t[i] - 6 *np.pi)
    return real, im

real, im = spiral(t)


plt.figure(dpi = 200)
plt.axis('equal')
plt.xlabel("Re")
plt.ylabel("Im")
plt.grid()
plt.plot(real,im,linewidth = 1)
plt.savefig("3a.png")
            
            
```

```python
t = np.linspace(0,2*np.pi,1000)
a = 1
b = 10

def gamma(t):
    return a * np.exp(1j*t) +  b * np.exp(-1j*t)


plt.figure(dpi = 200)
plt.title("Plot fr a=1 und b=10")
plt.axis('equal')
plt.xlabel("Re")
plt.ylabel("Im")
plt.grid()
plt.plot(gamma(t).real, gamma(t).imag,linewidth = 1)
plt.savefig("4a.png")
```
