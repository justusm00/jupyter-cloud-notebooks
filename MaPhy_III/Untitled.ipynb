```python
##1a

import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0,2*np.pi,1000)
n = np.array([1,3,5,7])

def f(x,n):
    a = np.zeros(len(x))
    for i in range(n):
        a += np.sin((2*i+1)*x)/(2*i+1) 
    return 4/np.pi*a

def g(x):
    a = np.zeros(len(x))
    for i in range (len(x)):
        if x[i] < np.pi:
            a[i] = 1
        else:
            a[i] = -1
            
    return a
        


fig = plt.figure(dpi = 200)

plt.title("look at this graph")
plt.grid()
plt.ylabel("y")
plt.xlabel("x")
for i in range(len(n)):
    plt.plot(x,f(x,i+1),linewidth= 1, label = "$f_%d$" %n[i])
plt.plot(x,g(x),linewidth= 1, label ="f")
plt.legend(loc=1) 
plt.savefig("1a.png")

        
```
