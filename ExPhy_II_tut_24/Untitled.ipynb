```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r1 = 1
r2 = 5
r3 = 10
eps1 = 1.5
eps2 = 4.5
```

```python
r = np.linspace(0, r3, 100)
def E_field(r):
    if r < r1:
        return 0
    elif r < r2:
        return 1/(eps1 * r**2)
    else:
        return 1/(eps2 * r**2)


def D_field(r):
    if r < r1:
        return 0
    else:
        return 1/r**2
```

```python
E = [E_field(r_val) for r_val in r]
D = [D_field(r_val) for r_val in r]
```

```python
fig, axs = plt.subplots(1,2,dpi =150)
axs = axs.flatten()
axs[0].plot(r,E,lw=1)
axs[0].set_title("E-Feld", fontsize=14)
axs[1].set_title("D-Feld", fontsize=14)
axs[0].set_xlabel("$r$ in cm", fontsize=14)
axs[1].set_xlabel("$r$ in cm", fontsize=14)
axs[0].set_ylabel(r"$E\cdot4\pi\epsilon_0/ Q$", fontsize=14)
axs[1].set_ylabel(r"$D\cdot4\pi/ Q$", fontsize=14)
axs[0].set_ylim(-0.01, 0.2)

axs[1].plot(r,D,lw=1)
#axs[1].set_ylim(0, 0.1)
plt.tight_layout()
plt.savefig("6_1b.pdf")
```
