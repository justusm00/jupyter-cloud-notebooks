```python
import numpy as np
import matplotlib.pyplot as plt

n=64 #number of solvent molecules
T=250 #temperature

fix,ax = plt.subplots(2,2,figsize=(12,8),dpi=100)
ax = ax.reshape(4)

for i in range(4):
    j = i+1
    t,v,vcst = np.loadtxt("clo{}_{}w_{}K.dt3".format(j,n,T),usecols=(1,4,11)).T
    vrel = np.abs(vcst/v)
    ax[i].set_title("clo{}_{}w_{}K".format(j,n,T))
    ax[i].set_xlabel("$t$ [fs]")
    ax[i].set_ylabel("$V_{cst}/V$")
    plt.tight_layout()
    ax[i].plot(t,vrel,lw=1)
    #plt.savefig("clo{}_{}w_{}K.png".format(i,n,T))
```
