```python
import numpy as np
import matplotlib.pyplot as plt

#compare moments of inertia over time with and without restraining potential


moi = np.loadtxt("clo_32w_250K.dt3", usecols = (6,7,8)).T  
RG= np.loadtxt("clo_32w_250K.dt3", usecols = 5)     #radius of gyration
moi_restr = np.loadtxt("clo_32w_250K_restr.dt3", usecols = (6,7,8)).T
RG_restr= np.loadtxt("clo_32w_250K_restr.dt3", usecols = 5)
t1 = np.loadtxt("clo_32w_250K.dt3", usecols = 1)
t2= np.loadtxt("clo_32w_250K_restr.dt3", usecols = 1)

#plot moments of inertia
colors = ['r','g','b']
fix1,ax1 = plt.subplots(1,dpi=100)
ax1.set_title("Moments of inertia")
ax1.set_xlabel("$t$ [fs]")
ax1.set_ylabel("$I$ [unit]")
plt.tight_layout()
for i in range(3):
    ax1.plot(t1,moi[i],'--',lw=1,color=colors[i],label=r"$I_{},V_c=0$".format(i))
    ax1.plot(t2,moi_restr[i],lw=1,color=colors[i],label=r"$I_{},V_c \neq 0$".format(i))
ax1.legend()
    
    
    
    
#plot radii of gyration
fix3,ax3 = plt.subplots(1,dpi=100)
ax3.set_title("Radius of gyration")
ax3.set_xlabel("$t$ [fs]")
ax3.set_ylabel("$R_G$ $[\AA]$")
plt.tight_layout()
ax3.plot(t1,RG,'r--',lw=1,label=r"$V_c=0$")
ax3.plot(t2,RG_restr,'b',lw=1,label=r"$V_c \neq 0$")
ax3.legend()


```
