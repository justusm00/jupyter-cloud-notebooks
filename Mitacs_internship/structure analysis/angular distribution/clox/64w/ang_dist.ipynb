```python
import numpy as np
import matplotlib.pyplot as plt

n=64 #number of solvent molecules
T=240 #temperature

fix,ax = plt.subplots(2,2,figsize=(12,8),dpi=150)
ax = ax.reshape(4)

for i in range(4):
    j = i+1
    a,p = np.loadtxt("clo{}_{}w_{}K.26".format(j,n,T),usecols=(0,1)).T
    ax[i].set_title("$x$={}".format(j))
    ax[i].set_xlabel(r"$r$ [$\AA$]")
    ax[i].set_ylabel("$p(r)$")
    plt.tight_layout()
    ax[i].plot(a,p,lw=1)
plt.savefig("clox_64w.png")
    

```
