```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x-2)**3-5*x**2

def rest(x,a):
    return f(a)+0.5*100*(x-a)**2

x = np.linspace(0,10,1000)
xrest=np.linspace(-1,1,100)
plt.figure(dpi=200)
plt.plot(x,f(x),lw=1)
plt.ylabel("Free Energy")
plt.xlabel("$r_{com}$")
#plt.xticks([])
plt.yticks([])
for i in range(1,10):
    plt.plot(xrest+i,rest(xrest+i,i),'r',lw=1)
plt.savefig("umbrella.png")
```
