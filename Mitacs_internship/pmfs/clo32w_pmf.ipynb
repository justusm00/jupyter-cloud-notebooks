```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

r = np.loadtxt("clo32w_pmf.dat",usecols=0)
p = np.loadtxt("clo32w_pmf.dat",usecols=3)

def f(r,a,b):
    return a*r**2*np.exp(-b*r**2)

p0=[0.0,0.0]
params,pcov = curve_fit(f,r,p,p0)
rfit=np.linspace(0,np.max(r),1000)

plt.plot(r,p)
plt.plot(rfit,f(rfit,params[0],params[1]))
```
