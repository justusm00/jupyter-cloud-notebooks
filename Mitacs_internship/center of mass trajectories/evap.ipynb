```python
import numpy as np
import matplotlib.pyplot as plt

##script to plot rcms trajectories of evaporating molecules

input_file="clo4_64w_250K.dt27"

N = 64 #number of water molecules
rmax = 13.6 #cms distance that defines evaporation


#load data
rcom = np.loadtxt(input_file,usecols=1)
#vrel = np.loadtxt(input_file,usecols=2)
nsteps = int(len(rcom)/N)
rcom = rcom.reshape(int(len(rcom)/N),N)
#vrel = vrel.reshape(int(len(vrel)/N),N)


tmax = nsteps*10

#define time array
t = np.linspace(0,tmax,nsteps)

#average center of mass distance averaged over water molecules
rmeans = np.array([])
for i in range(N):
    rmeans = np.append(np.mean(rcom[:,i]),rmeans)
    
rmean = np.mean(rmeans)
print(rmean)

#plot trajectories of evaporating molecules
for i in range(N):
    if max(rcom[:,i])>rmax:
        plt.figure(dpi=100)
        plt.xlabel("$t$ [fs]")
        plt.ylabel("$r_{water-com}$ $[\AA]$")
        plt.title("{}, water molecule #{} ".format(input_file[:-5],i))
        plt.plot(t,rcom[:,i],lw=1)
        
        
        

```

```python
j=27
#plot and save best trajectory
plt.figure(dpi=100)
plt.xlabel("$t$ [fs]")
plt.ylabel("$r_{water-com}$ $[\AA]$")
plt.title("{}, water molecule #{} ".format(input_file[:-5],j))
plt.plot(t,rcom[:,j],lw=1)
plt.savefig("{}_evap.png".format(input_file[:-5]))
```
