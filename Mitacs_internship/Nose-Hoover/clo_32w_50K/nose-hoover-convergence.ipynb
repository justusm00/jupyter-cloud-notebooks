```python
##temperature averaged over 100fs

import numpy as np
import matplotlib.pyplot as plt

T=50

t,Tavr = np.loadtxt("clo_32w_50K.dt3",usecols=(1,3)).T #time and average temperature (averaged over 10fs)
nsteps = len(t)

#re-average over 100fs intervals
Tavrr= np.array([np.mean(Tavr[1:10])])
for i in range(10,nsteps,10):
    Tavrr = np.append(np.mean(Tavr[i:(i+10)]),Tavrr)
    

plt.figure(dpi=150)
#plt.xlim(0,2000)
plt.title("Temperature averaged over 100fs intervals, $T_0$={}K".format(T))
plt.scatter(t[::10],Tavrr,marker='.',s=10,c='b')
plt.hlines(np.mean(Tavr[1:]),t[0],t[-1],color='r',lw=1)
plt.xlabel("$t$ [fs]")
plt.ylabel(r"$\bar{T}$ [K]")
plt.savefig("clo_32w_50K_tavr.png")


```
