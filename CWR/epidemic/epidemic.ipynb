```python
import numpy as np
import matplotlib.pyplot as plt

N = 83e6
t1 = np.loadtxt("epidemic1.csv", usecols = 0)
S1 = np.loadtxt("epidemic1.csv", usecols = 1) / N * 100
E1 = np.loadtxt("epidemic1.csv", usecols = 2) / N * 100
I1 = np.loadtxt("epidemic1.csv", usecols = 3) / N * 100
R1 = 100 - S1 - E1 - I1

S2 = np.loadtxt("epidemic2.csv", usecols = 1) / N * 100
E2 = np.loadtxt("epidemic2.csv", usecols = 2) / N * 100
I2 = np.loadtxt("epidemic2.csv", usecols = 3) / N * 100
R2 = 100 - S2 - E2 - I2

S3 = np.loadtxt("epidemic3.csv", usecols = 1) / N * 100 
E3 = np.loadtxt("epidemic3.csv", usecols = 2) / N * 100
I3 = np.loadtxt("epidemic3.csv", usecols = 3) / N * 100
R3 = 100 - S3 - E3 - I3

fig, axs = plt.subplots(4, figsize=(12,14), dpi = 200)
plt.subplots_adjust(hspace = 0.5)



axs[0].grid()
axs[0].set_ylabel("percentage of population")
axs[0].set_xlabel("days passed since the outbreak")
axs[0].plot(t1, S1, lw = 1, label = "$R_0=1.25$")
axs[0].plot(t1, S2, lw = 1, label = "$R_0=1.5$")
axs[0].plot(t1, S3, lw = 1, label = "$R_0=2$")
axs[0].legend(loc = "upper right")
axs[0].set_title("$S$-group")



axs[1].grid()
axs[1].set_xlabel("days passed since the outbreak")
axs[1].set_ylabel("percentage of population")
axs[1].plot(t1, E1, lw = 1, label = "$R_0=1.25$")
axs[1].plot(t1, E2, lw = 1, label = "$R_0=1.5$")
axs[1].plot(t1, E3, lw = 1, label = "$R_0=2$")
axs[1].legend(loc = "upper right")
axs[1].set_title("$E$-group")


axs[2].grid()
axs[2].set_xlabel("days passed since the outbreak")
axs[2].set_ylabel("percentage of population")
axs[2].plot(t1, I1, lw = 1, label = "$R_0=1.25$")
axs[2].plot(t1, I2, lw = 1, label = "$R_0=1.5$")
axs[2].plot(t1, I3, lw = 1, label = "$R_0=2$")
axs[2].legend(loc = "upper right")
axs[2].set_title("$I$-group")



axs[3].grid()
axs[3].set_xlabel("days passed since the outbreak")
axs[3].set_ylabel("percentage of population")
axs[3].plot(t1, R1, lw = 1, label = "$R_0=1.25$")
axs[3].plot(t1, R2, lw = 1, label = "$R_0=1.5$")
axs[3].plot(t1, R3, lw = 1, label = "$R_0=2$")
axs[3].legend(loc = "upper right")
axs[3].set_title("$R$-group")


fig.savefig("epidemic.png")
```
