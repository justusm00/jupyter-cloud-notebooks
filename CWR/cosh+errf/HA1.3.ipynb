```python
import numpy as np
import matplotlib.pyplot as plt

##erf(x) 
x = np.loadtxt("data3.txt", usecols = 0)
y = np.loadtxt("data3.txt", usecols = 1)



plt.figure(dpi=400)
plt.title("plot of the error function")
plt.grid()
plt.ylabel("erf(x)")
plt.xlabel("x")
plt.plot(x,y,lw=1)
```

```python
##cosh und Fehler

x = np.loadtxt("cosh.txt", usecols = 0)  ##delta x
simp_y = np.loadtxt("cosh.txt", usecols = 1)  ### integral über coshx
mid_y = np.loadtxt("cosh.txt", usecols = 2)

value = np.sinh(1) ##tatsächlicher Wert des Integrals
simp_err = np.abs(simp_y - value)  ##Fehler
mid_err = np.abs(mid_y - value)


plt.figure(dpi=200)
plt.yscale('log')
plt.xscale('log')
plt.title("error of cosh(x)")
plt.grid()
plt.ylabel("$\Delta$")
plt.xlabel("$\Delta x$")
plt.plot(x,simp_y,lw=1,label = "simpson rule")
plt.plot(x,mid_y,lw=1,label = "midpoint rule")
plt.legend(loc ="upper left")
plt.savefig("cosh.png")
```
