```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
import scipy.stats as sci

res = 3
riem = np.abs(3 - np.loadtxt("data1.txt")) ##riemannsumme
trap = np.abs(3 - np.loadtxt("data2.txt")) ##trapezregel
N = np.linspace(1,7,7)
slope1 =  sci.stats.linregress(N, np.log10(riem))
slope2 =  sci.stats.linregress(N, np.log10(trap))

plt.figure(dpi=200)
plt.title("double logarithmic plot of the absolute errors")
plt.grid()
plt.xlabel("$\log(N)$")
plt.ylabel("$\log(\Delta)$")
plt.plot(N,np.log10(riem),lw=1,label = "riemann sum")
plt.plot(N,np.log10(trap),lw=1, label = "trapezoid rule")
plt.legend(loc = "upper right")

print(slope1)
print(slope2)
```
