```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint



def limit(y,t):
    a = y[0]
    b = y[1]
    da = a-b-a*(a**2+5*b**2)
    db = a + b - b*(a**2+b**2)
    return [da,db]

t = np.linspace(0,100,1000) #time
x0 = np.linspace(0,1,10)
y0 = np.linspace(0,1,10)


fig, ax = plt.subplots(dpi=100)
rmin = plt.Circle((0,0),1/np.sqrt(2),fill=False,color="r")
rmax = plt.Circle((0,0),1,fill=False,color="r")
ax.add_patch(rmin)
ax.add_patch(rmax)
for x0,y0 in zip(x0,y0):
    y = odeint(limit, [x0,y0], t).T
    ax.plot(y[0],y[1],lw=1,color="b")
```

```python
import numpy as np 
import matplotlib.pyplot as plt


## HERE PARAMETERS

a = 1
b = 1
c = 1
lam = 1
grid = 20


x = np.linspace(-5,5,grid)
y = np.linspace(-5,5,grid)
rx,ry = np.meshgrid(x,y)


### HERE SYSTEM

def bla(x,y):
    dx = a*x*(b-x**2)
    dy = c
    return np.array([dx,dy])

flowx, flowy = bla(rx,ry)

plt.figure(dpi=100)
plt.title("flow")
N = np.sqrt(flowx**2 + flowy**2)
N[N==0] = np.nan
plt.quiver(rx, ry, flowx/N, flowy/N, np.log(N))
plt.xlabel("x")
plt.ylabel(r"$\nu$")
```
