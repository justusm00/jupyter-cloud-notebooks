```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root
from scipy.linalg import eig
```

```python
##define dimensionless system
def hopf(y,t,mu):
    x = y[0]
    y = y[1]
    dxdt = -y + mu*x +x*y**2
    dydt = x + mu*y - x**2
    return [dxdt, dydt] 

#define jacobian
def hopf_jac(y,t,mu):
    x = y[0]
    y = y[1]
    dxdx = mu + y**2
    dxdy = -1 +2*x*y
    dydx = 1 - 2*x
    dydy = mu
    return np.array([[dxdx,dxdy],[dydx,dydy]])
```

```python
samples = 20

#set a,b ranges
x = np.linspace(-1, 2, samples)
y = np.linspace( -1.5, 1.5, samples)
X,Y = np.meshgrid(x, y)
mus = np.array([-1,-0.5,0.4])


for mu in mus:
    plt.figure(dpi=100)
    DX,DY = hopf([X,Y], 0, mu)
    nrm = np.sqrt(DX**2 + DY**2)
    plt.streamplot(X, Y, DX/nrm, DY/nrm)
    plt.title("$\mu={}$".format(mu))
    plt.xlabel("x")
    plt.ylabel("y")
```

We can see that for $\mu>0$ the fixed point at (x,y)=(0,0) becomes unstable and there is a stable limit cycle somewhere, probably...
