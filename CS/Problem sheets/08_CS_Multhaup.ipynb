```python
#constants
sigma = 10
b = 8/3


##define Lorenz system
def lorenz(y,t):
    x = y[0]
    z = y[2]
    y = y[1]
    dxdt = sigma*(y-x)
    dydt = r*x-y-x*z
    dzdt = x*y-b*z
    return [dxdt, dydt ,dzdt] 


#define jacobian
def lorenz_jac(y,t):
    x = y[0]
    z = y[2]
    y = y[1]
    dxdx = -sigma
    dxdy = sigma
    dxdz = 0
    dydx = r - z
    dydy = -1
    dydz = -x
    dzdx = y
    dzdy = x
    dzdz = -b
    return np.array([[dxdx,dxdy,dxdz],[dydx,dydy,dydz],[dzdx,dzdy,dzdz]])
```

```python
#plot trajectories in 3d
r = 28

fig = plt.figure(dpi=100)
ax = fig.gca(projection='3d')


y0 = [0, 1,1.05]
t = np.linspace(0,50,10000)
y = odeint(lorenz, y0, t)
ax.set_title("$r={}$".format(r))
ax.plot(y[:,0],y[:,1],y[:,2],lw = 1)

##plot individual coordinates
fig, ax = plt.subplots(3,dpi=100)
plt.tight_layout()
for i in range(3):
    ax[i].plot(t,y[:,i])
ax[2].set_xlabel("t")
ax[0].set_ylabel("x")
ax[1].set_ylabel("y")
ax[2].set_ylabel("z")
```
