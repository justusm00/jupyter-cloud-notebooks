```python
import numpy as np
import matplotlib.pyplot as plt

dt = 1 #time step
tmax = 100 ##max time
N = 100 #no of walkers
steps = np.random.randint(2, size=(int(tmax/dt),N)) * 2 - 1 ##walker steps
position = np.cumsum(steps,axis= 0) #walker postions

rms = np.zeros(int(tmax/dt))

#calculate rms
for i in range(len(rms)):
    rms[i] = 1/N * np.sum(position[i]**2)

#analytical rms
def rms_ana(t):
    return t/dt

t = np.arange(0,tmax,dt)

plt.figure(dpi=100)
plt.title("RMS of random walk as function of time")
plt.xlabel("$t$")
plt.ylabel("RMS")
plt.plot(t,rms_ana(t),'r',label = "analytical")
plt.scatter(t,rms,label = "numerical")
plt.legend(loc=0)
```

We can see that the numerical calculations show the expected linear behaviour.
