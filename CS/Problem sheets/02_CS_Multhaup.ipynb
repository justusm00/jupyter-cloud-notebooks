```python
import numpy as np 
import matplotlib.pyplot as plt

gamma = 1
omega = 1.5
grid = 20


x = np.linspace(-10,10,grid)
y = np.linspace(-10,10,grid)
rx,ry = np.meshgrid(x,y)


def bla(x,y):
    return y, -2*gamma*y - omega**2*x

flowx, flowy = bla(rx,ry)

plt.figure(dpi=100)
plt.title("flow for $\gamma={}$ and $\omega={}$".format(gamma,omega))
N = np.sqrt(flowx**2 + flowy**2)
N[N==0] = np.nan
plt.quiver(rx, ry, flowx/N, flowy/N, np.log(N))
plt.xlabel("x")
plt.ylabel(r"$\nu$")
```

We can see that there is a fixed point at $(x,\nu)=(0,0)\,$, which is also easily seen from the equations. As the vector field curls around this point, the fixed point is stable for $\gamma<\omega$.

```python
import numpy as np 
import matplotlib.pyplot as plt

gamma = 2
omega = 1.5
grid = 20


x = np.linspace(-1,1,grid)
y = np.linspace(-1,1,grid)
rx,ry = np.meshgrid(x,y)


def bla(x,y):
    return y, -2*gamma*y - omega**2*x

flowx, flowy = bla(rx,ry)

plt.figure(dpi=100)
plt.title("flow for $\gamma={}$ and $\omega={}$".format(gamma,omega))
N = np.sqrt(flowx**2 + flowy**2)
N[N==0] = np.nan
plt.quiver(rx, ry, flowx/N, flowy/N, np.log(N))
plt.xlabel("x")
plt.ylabel(r"$\nu$")
```

Again there is a fixed point at $(x,\nu)=(0,0)\,$, only that this time there is always some flow pointing away in a specific direction, therefore for $\gamma>\omega$ we get some kind of unstable saddle.

```python
import numpy as np 
import matplotlib.pyplot as plt


##first the x system
r = 10
x = np.linspace(0,5,1000)

def xdot(x):
    return x*(r-np.exp(x))

plt.figure(dpi=100)
plt.grid()
plt.title("phase space plot for $r={}$".format(r))
plt.plot(x,xdot(x))
plt.ylabel("$\dot x$")
plt.xlabel("$x$")
```

```python
##now the y system
xi = -5
y = np.linspace(-3,3,1000)

def ydot(x):
    return y**4-4*y + xi

plt.figure(dpi=100)
plt.grid()
plt.title(r"phase space plot for $\xi={}$".format(xi))
plt.plot(y,ydot(y))
plt.ylabel("$\dot y$")
plt.xlabel("$y$")
```

3: Overfishing

```python
K = 1e6
N0 = 1e5
mu = 0.5
Hc = mu*K/4

H = np.linspace(0,Hc,1000)

##steady state solution = stable fixed point 
def Nst(H):
    return K/2 + np.sqrt(K**2/4 - H*K/mu)

plt.figure(dpi=100)
plt.grid()
plt.title("steady state population as a function of H")
plt.plot(H,Nst(H)/K,label = "population",color = "blue")
plt.scatter(Hc,Nst(Hc)/K,label = "$H_c$",color = "red")
plt.legend(loc=1)
plt.ylabel("$N_{st}(H)/K$")
plt.xlabel("$H$")
```

For $H>H_c$, the stable fixed point becomes non-real and therefore the population collapses.
