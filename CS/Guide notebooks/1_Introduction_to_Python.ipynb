### Conditional execution

Like in other languages, you can use `if`-statements to control the execution of code. The following code demonstrates the syntax for this. The expression after the `if` keyword is evaluated first. If it's reduced to a *logical* "true", the following indented code is executed, otherwise it's bypassed. Check that only the first print command is executed:

```python
a = 1
b = 2

if a<b:
    print(f"{a} is smaller than {b}")
    
if (a+b)<3:
    print(f"{a}+{b} is smaller than 3")
```

With the `else`-keyword you can provide alternative code, that is executed in case the previous `if`-statement was bypassed. The following code checks a list for multiples of 3:

```python
myList = [1, 2, 3, 6, 9, 10, 11, 12]

for x in myList:
    if x%3 == 0:
        print(f"{x} is divisible by 3.")
    else:
        print(f"{x} is not divisible by 3.")
```

### Defining functions

You can use the keyword `def` to define you own functions in Python. In the simplest of cases, a function will take a number of parameters, calculate new data from that and return the result.

The following function squares a number:

```python
def square(x):
    return x*x

print(square(4))
```

The following functions extracts all numbers from a given list that are multiples of a second parameter:

```python
def multiples(oldList, divisor):
    newList = []
    for x in oldList:
        if x%divisor == 0:
            newList.append(x)
    return newList

myList = [1, 2, 3, 6, 9, 10, 11, 12]

MultiplesOfThree = multiples(myList, 3)

print(MultiplesOfThree)
```

### Function scope

Functions have a peculiar behaviour in regards to reading and writing names, **that you should know about**.

* A function can read from all global names
* A function can (by default) only bind to local names

Study the following code:

```python
def foo():
    print("foo: ", i)
    
def bar():
    i = 3
    print("bar: ", i)

    
i = 2
    
print(i)    # 2

foo()       # 2
bar()       # 3

print(i)    # 2
```

We set the name `i=2` and check it's content with the first print command. After that, the function `foo()` accesses `i` and prints out it's value 2 as well.

Function `bar()` binds the value 3 to the name `i` and prints it out accordingly. However, the last `print()` prints out 2 again. That's because the command `i=3` in `bar()` created a local name `i` that masks the global `i`.

If you want to have `bar()` actually bind an object to *global* name `i`, you have to state this with the keyword `global`:

```python
def foo():
    print("foo: ", i)
    
def bar():
    global i
    i = 3
    print("bar: ", i)

    
i = 2
    
print(i)    # 2

foo()       # 2
bar()       # 3

print(i)    # 3
```

An extremely useful way of generating 2D arrays are *meshgrids*. Meshgrid constructs two 2D arrays from two 1D arrays, where one of the 1D arrays is interpreted as the x-axis and the other as the y-axis.

Let's first look at an example:

```python
x = [0, 1, 2]
y = [5, 6, 7]

X,Y = np.meshgrid(x,y)

print(X,"\n")
print(Y)
```

As cou can see, each of the two 1D arrays is copied along a different axis. Let's create a square calculation domain of size 10x10, with 200 samples per axis:

```python
x = np.linspace(-5, 5, 200)
X,Y = np.meshgrid(x,x)
```

Now lets calculate the function $\sin(x)\cdot \cos(y)$ on this domain. **Take your time** to understand the way this calculation works and have a look at the actual arrays involved if necessary. Remember that everything happens element-wise:

```python
Z = np.sin(X) * np.cos(Y)
plt.imshow(Z)
```

A more intricate function:

$$ \cos(5r) \cdot e^{-r} $$

```python
R = np.sqrt( X**2 + Y**2)
Z = np.cos(5*R) * np.exp(-R)

plt.imshow(Z)
```

These calculations are called **vectorized**. Unlike in languages like *C*, we did not explicitly iterate over the entries of the array with a for loop, but just stated our high-level intention. One advantage of this is quite appealing and readable code. More important though, this gives us C-like performance in Python. NumPy's operations on arrays, such as multiplication or the square root, do not rely on Python code, but are instead compiled to efficient machine code, which actually runs fast enough to work with actual, real world data and not just little example sets.

As a good rule of thumb when working with tools like NumPy (or Matlab): **Avoid `for` loops at all costs for heavy number crunching**. Of course, you'll need `for`-loops for initializations, scripting of multiple runs and so on. But think carefully about any `for`-loop you are using. In most cases, you'll find that you do not need them.

Let's demonstrate the performance difference with a simple vector multiplication. First we create two large random vectors:

```python
v1 = np.random.rand(10_000_000)
v2 = np.random.rand(10_000_000)
```

Now let's do a vector multiplication by hand and measure the runtime:

```python
import time

result = 0
begin = time.time() # Get the current time

for i in range(10_000_000):
    result = result + v1[i]*v2[i]

delta1 = time.time() - begin # Get the current time and compare it to before
    
print("Time elapsed: ", delta1)
```

Now with NumPy's `dot()` method:

```python
begin = time.time()

result = v1.dot(v2)

delta2 = time.time() - begin

print("Time elapsed: ", delta2)

print(f"NumPy was {delta1/delta2} times faster.")
```

You should see a performance increase of at least 2-3 magnitudes with this simple operation.

### Ordinary Differential Equations

You can use SciPy to numerically solve ordinary differential equations. Take for example the following Lotka-Volterra equations:

\begin{align}
\dot{x} & = \frac{2}{3}x - \frac{4}{3}xy \\
\dot{y} & = xy-y
\end{align}

First, import the `odeint` function from SciPy. `odeint` is a very easy to use interface to SciPy's integrators with lots of sensible default values. It's suitable for a quick test, but in general you'll want to have a deeper look into your integrators if you are doing any serious calculations! We will cover this in more detail at a later time.

```python
from scipy.integrate import odeint
```

Then, define the system function (the right hand side of the ODE system above) as a Python function. The first parameter to the function is a list containing all the state variables of your system.

```python
def lotka(y, t):
    dx = 2/3 * y[0] - 4/3 * y[0]*y[1]
    dy = y[0]*y[1] - y[1]
    
    return [dx, dy]
```

Create a list *y0* of initial conditions and an array of *times* at which you want to know the solution to the problem. Note that this is **not** related to the time-step used by the integrator (which is controlled automatically by default):

```python
y0 = [1, 1]
t = np.linspace(0,20,100)
```

Finally, call the `odeint()` function and obtain the solution.

```python
y = odeint(lotka, y0, t)
plt.plot(t,y)
```

If you want just the data of a single independent variable, you can use the colon-operator `:` to slice y:

```python
plt.plot(t,y[:,0])
```

`y[:,0]` means that we want to extract *entire* 0'th column.  For more details on slicing, please refer to the NumPy documentation:

https://docs.scipy.org/doc/numpy/reference/arrays.indexing.html

Let's try to visualize the phase space. For this we generate a couple of initial conditions between 0 and 2 and solve the trajectories, which we plot into the $x-y$ phase space. We access the $x$ and $y$ variable by slicing the `y`-array into its separate columns.

```python
t = np.linspace(0,15,200)

for i in range(30):
    y0 = np.random.rand(2)*2
    y = odeint(lotka, y0, t)
    plt.plot(y[:,0], y[:,1])
```

Lastly, let's draw a portrait of the phase space around the fixpoint. Use `meshgrid()` to generate 20x20 sample points:

```python
x = np.linspace(0,5,20)
y = np.linspace(0,3,20)

X,Y = np.meshgrid(x,y)
```

Then, calculate the differentials at these sample points:

```python
DX = 2/3*X - 4/3*X*Y
DY = X*Y - Y
```

Finally, draw the portrait with the `quiver()` function. Use the `xlabel` and `ylabel` functions to label the axis:

```python
plt.quiver(X,Y,DX,DY)
plt.xlabel("x")
plt.ylabel("y")
```

It might be better to normalize the vectors and color them depending on the flow strength. We remove the zeros from the normalization and substitute them with `NaN`s. This will make Matplotlib ignore those values.



```python
N = np.sqrt(DX**2 + DY**2)
N[N==0] = np.nan
plt.quiver(X, Y, DX/N, DY/N, np.log(N))
plt.xlabel("x")
plt.ylabel("y")
```
