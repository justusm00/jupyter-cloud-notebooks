### Problem 1: Hair cell oscillations

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#Parameters 
#(forces in pN, distances in nm, times arbitrary)

alpha=0.0017
beta= 432
delta= - 1/4
epsilon = .01
eta = 1


# definition of flow

def flow(y, t):
    dx = - alpha * ( ( y[0] - y[1] )**3 - beta * ( y[0] - y[1] ) ) - delta * y[1]
    dy =  epsilon * ( y[1]  -  eta * y[0] )  
    return [dx, dy]
```

```python
t = np.linspace(0,800,4000)
y0 = (-5, 1)

y = odeint(flow, y0, t)

plt.figure(figsize=(15,4))
plt.plot(t, y[:,0])
plt.xlabel("time")
plt.ylabel("x");
#plt.axes().set_aspect(0.2)
```

```python
t = np.linspace(0,400,4000)
x = np.linspace(-1, 1)
X,Y = np.meshgrid(x,x)

plt.figure(figsize=(5,5))    
for i in range(3):  
    y0 = 0,1+i
    y = odeint(flow, y0,t)
    plt.plot(y[:,0], y[:,1])

plt.xlabel("x")
plt.ylabel("y");
```
