```python
import numpy as np
import matplotlib.pyplot as plt

R = 5
h0 = 10
g = 9.81
r = np.arange(-R,R,0.001)
w1 = 1
w2 = 2
w3 = 3
def fluidSurface(r,w):
    return h0 - w**2*R**2/(4*g) + w**2*r**2/(2*g) 
l = np.array([])
for i in range (len(r)):
    l = np.append(h0,l)

plt.figure()
plt.grid(True)
plt.title("Oberfläche des Quecksilbers")
plt.xlabel("Abstand r vom Mittelpunkt")
plt.ylabel("Höhe z")
plt.plot(r,fluidSurface(r,w1), label = "$\omega_1$")
plt.plot(r,fluidSurface(r,w2), label = "$\omega_2$")
plt.plot(r,fluidSurface(r,w3), label = "$\omega_3$")
plt.plot(r,l, label = 'h$_0$')
plt.legend()

```
