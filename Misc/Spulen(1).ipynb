```python
import matplotlib.pyplot as plt
import math
import numpy as np
import scipy.optimize as opt

plt.style.use('default')
fig, ax = plt.subplots()

data = np.loadtxt("InduLS.txt")
d   = data[:,0]*(10**(-2))
x   = data[:,1]
Ri  = 10*(10**3)
k   = 5.109*(10**(-10))
n   = 369
A   = np.pi*((0.059)/2)**2

B = ((Ri*k*x)/(n*A))*10**3

data2 = np.loadtxt("HallLS.txt")
d2 = data2[:,0]*(10**(-2))
x2 = data2[:,1]*(10**(-1))


plt.errorbar(d,B,0.00000000005,0.00000000035,fmt='none',capsize=3,label='Messwerte', color="orange")
plt.errorbar(d2,x2,0.00000000005,0.00000000035,fmt='none',capsize=3,label='Messwerte 2', color="blue")


plt.xlabel("d in [m]")
plt.ylabel("B in [mT]")

plt.grid(True)
ax.legend()
plt.savefig("SpuleL.pdf")
plt.show()
print(A,B)
```

```python
d3 = -data2[:,0]*(10**(-2))
x3 = data2[:,1]*(10**(-1))




plt.errorbar(d3,x3,0.00000000005,0.00000000035,fmt='none',capsize=3,label='Messwerte lange Spule', color="blue")
plt.errorbar(d2,x2,0.00000000005,0.00000000035,fmt='none',capsize=3, color="blue")


l  = 451 * (10**(-3))
I  = 0.49
r  = (128*(10**(-3)))/2
a  = np.arange (-0.25, 0.7, 0.01)
n2 = 831

mu = 1.25663706212*(10**(-6))

def H(I,r,l,a,n2):
    return (1/2)*((n2*I)/l)*((a/np.sqrt(r**2+a**2))+((l-a)/np.sqrt(r**2+(l-a)**2)))

plt.plot(a-0.2255, H(I,r,l,a,n)*mu*2500, label='theoretische Kurve')

th = mu * n2 * I/l * 1000

plt.plot(np.linspace(-0.2, 0.2, 30), np.linspace(th, th, 30), label='Theoretische Flussdichte', linestyle="--")

plt.xlabel("d in [m]")
plt.ylabel("B in [mT]")

plt.grid(True)
plt.legend()
plt.savefig("SpuleLT.pdf")
plt.show()
```
