```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

k = 0.1 ##friction
b = 0.35 ##driving amplitude
w = 1.4 #frequency


def duffing(y,t):
    dx = y[1]
    dy = b*np.cos(w*t) + y[0] - y[0]**3 - k*y[1]
    return [dx,dy]
    

y0 = [0,0]
t = np.linspace(0,200,10000)

y = odeint(duffing, y0, t).T

plt.figure(dpi=150)
plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.plot(y[0],y[1],lw=1)
```
