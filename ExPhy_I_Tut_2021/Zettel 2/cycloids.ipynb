```python
import numpy as np
import matplotlib.pyplot as plt

##define constants
omega = 1
R = 1

t = np.linspace(0,20,200) #time interval

##cycloid function
def cycloid(t):
    x = R*(omega*t+np.sin(omega*t))
    y = R*(1+np.cos(omega*t))
    return x,y

#cycloid velocity
def cycloid_v(t):
    vx = R*omega*(1+np.cos(omega*t))
    vy = -R*omega*np.sin(omega*t)
    return vx,vy

#cycloid acceleration
def cycloid_a(t):
    ax = R*omega**2*np.sin(omega*t)
    ay = -R*omega**2*np.cos(omega*t)
    return ax,ay

#data to plot
cycle = cycloid(t)
cycle_v = cycloid_v(t)
cycle_a = cycloid_a(t)

#plot trajectory, velocity and acceleration
fig, axs = plt.subplots(3, figsize= (7,7),dpi = 100)
plt.subplots_adjust(hspace = 0.5)
axs[0].grid()
axs[0].set_ylabel("$y$")
axs[0].set_xlabel("$x$")
axs[0].plot(cycle[0],cycle[1])

axs[1].grid()
axs[1].axis("equal")
axs[1].set_ylabel("$\dot{y}$")
axs[1].set_xlabel("$\dot{x}$")
axs[1].plot(cycle_v[0],cycle_v[1])

axs[2].grid()
axs[2].axis("equal")
axs[2].set_ylabel("$\ddot{y}$")
axs[2].set_xlabel("$\ddot{x}$")
axs[2].plot(cycle_a[0],cycle_a[1])
```
