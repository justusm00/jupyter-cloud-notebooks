## Gleichförmig beschleunigte, geradlinige Bewegung

In dieser Aufgabe werden wir die Bahnkurve eines gleichförmig beschleunigten Objektes in einer Dimension berechnen. Die Position $x$ zum Zeitpunkt $t$ ist gegeben durch folgende Gleichung:
\begin{equation*}
x\!\left( t \right) = x_0 + v_0 t + \frac{1}{2} a t^2, 
\end{equation*}
wobei $x_0$ und $v_0$ die Anfangsposition und -geschwindigkeit sind und $a$ die konstante Erdbeschleunigung, die auf das Objekt wirkt.

Schreiben Sie hier explizit die oben genannte Gleichung zur Berechnung von jedem Wert hin.


````python
t_1 = 1
t_2 = 2
t_3 = 5

x_0 = 3 #Anfangsposition
v_0 = 10 #Anfangsgeschwindigkeit
a = g = -9.81 #Erdbeschleunigung

x_1 = x_0 + v_0*t_1 + 1/2 * a * t_1**2 #Formel für Bahnkurvenpunkt x_1
x_2 = x_0 + v_0*t_2 + 1/2 * a * t_2**2 #Formel für Bahnkurvenpunkt x_2
x_3 = x_0 + v_0*t_3 + 1/2 * a * t_3**2 #Formel für Bahnkurvenpunkt x_3

print(x_1, x_2, x_3) #Ausgabe der Positionen der Bahnkurvenpunkte

````
