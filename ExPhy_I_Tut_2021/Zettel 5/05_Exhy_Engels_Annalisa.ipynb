## Integration einer Bewegungsgleichung: Freier Fall - Wiederholung

In dieser Aufgabe werden wir erneut die Bewegungsgleichung des freien Falls numerisch mit dem Euler Verfahren lösen. Die Bewegungsgleichung hat die Form:
\begin{equation}\tag{1}
 \ddot x = \frac{\mathrm{d}^2x}{\mathrm{d}t^2}= a
\end{equation}

Dieses Mal soll die Bewegungsgleichung nicht "per Hand" sondern rekursiv mittels einer Schleife gelöst werden. 

## 1. Integration mit Schleife

**a)** Lösen Sie erneut Gleichung (1) für fünf Zeitschritte $\Delta t = 0.1$, indem Sie die Gleichungen (3) und (4) für die Anfangswerte $x_0=x\left(t=0\right) = 2$ und $v_0=v\left(t=0\right) = 1$ numerisch mit Hilfe einer Schleife integrieren. Dabei soll $a=-9.81$ sein. Speichern Sie die numerisch berechneten Positionen in einer Liste indem Sie `append()` benutzen.

Als Erinerrung:

\begin{equation} \tag{2}
x\!\left( t \right) = x_0 + v_0 t + \frac{1}{2} a t^2.
\end{equation}

\begin{align}\tag{3}
        x\left(t+\Delta t \right) &= x\left(t\right) + v\left(t\right) \Delta t \\ 
        v\left(t+\Delta t \right) &= v\left(t\right) + a\Delta t \tag{4}
    \end{align}
    


```python
#Numpy array importieren
import numpy as np

#Definieren der Anfangswerte
x0 = 2
v0 = 1
deltat = 0.1
a = -9.81

#Berechnen der Positionen x für die Bahnkurvenpunkte
liste = [] #leere Liste
for i in range (0, 5): #Schleife
    x = x0 + deltat* i  * (v0 + 0.5 *deltat *a*(i-1)) #Befehl für die Schleife
    liste.append(x) #Ergebnis der Rechnung an die Liste anhägen
print(x) #Endergebnis ausgeben
print(liste) #Liste der Zwischenergebnisse ausgeben
```

**b)** Plotten Sie diese Positionen in einem Diagramm mit der Bahnkurve (Gleichung (2)). Benutzen Sie das mit `ax.scatter()` erweiterte Grundgerüst von `helpPlotScatter.ipynb`. 

```python
import numpy as np
import matplotlib.pyplot as plt

#Numerische Löstung plotten:
 #Positionen x aus der Liste aus Aufgabe 1 auf der y-Achse, Zeit auf der x-Achse

#Definieren von listet für die Zeitpunkte
deltat = 0.1
listet = [] #leere Liste
for i in range (0, 5): #Schleife
    t = deltat* i #Befehl für die Schleife
    listet.append(t) #Ergebnis der Rechnung an die Liste anhägen
print(listet) #Liste der Zwischenergebnisse ausgeben

#Diagramm darstellen
fig, ax = plt.subplots()
ax.set_title("Geradlinige Beschleunigung")  # Titel

#Numerische Lösung
ax.scatter(listet, liste, label='numerische Lösung') 

#Analytische Lösung
z = np.linspace(0, 4*deltat, 20)
xAnalytic = x0 + v0 * z + 0.5 * a * z**2

ax.plot(z, xAnalytic, label='analytische Lösung')

ax.set_xlabel("Zeit $t$")
ax.set_ylabel("Position $x$")

plt.legend(loc=3)
plt.show()
```

**a)** Wiederholen Sie Aufgabe 1 mit einem Zeitschritt $\Delta t = 0.005$. Achten Sie darauf, dass die Integrationszeit gleich bleibt! Da Sie nun sehr viele Werte erzeugt haben, ist es sinnvoller die numerische Lösung ebenfalls in einem Linienplot mit `ax.plot()` darzustellen.

```python
#Numpy array und matplotlib importieren
import numpy as np
import matplotlib.pyplot as plt

#Definieren der Anfangswerte
x0 = 2
v0 = 1
dt = 0.005
a = -9.81

#Diagramm darstellen
fig, ax = plt.subplots()
ax.set_title("Geradlinige Beschleunigung")  # Titel

#Numerische Löstung plotten:
 #Positionen x aus der Liste aus Aufgabe 1 auf der y-Achse, Zeit auf der x-Achse
#Berechnen der Positionen x für die Bahnkurvenpunkte
liste = [] #leere Liste
for i in range (0, 81): #Schleife
    x = x0 + dt* i  * (v0 + 0.5 *dt *a*(i-1)) #Befehl für die Schleife
    liste.append(x) #Ergebnis der Rechnung an die Liste anhägen
#Definieren von listet für die Zeitpunkte
listet = [] #leere Liste
for i in range (0, 81): #Schleife
    t = dt* i #Befehl für die Schleife
    listet.append(t) #Ergebnis der Rechnung an die Liste anhägen
print(t) #Endergebnis der Zeit ausgeben
#Numerische Lösung als Linienplot
ax.plot(listet, liste, label='numerische Lösung') 


#Analytische Lösung
z = np.linspace(0, 4*deltat, 20)
xAnalytic = x0 + v0 * z + 0.5 * a * z**2

ax.plot(z, xAnalytic, label='analytische Lösung')

ax.set_xlabel("Zeit $t$")
ax.set_ylabel("Position $x$")

plt.legend(loc=3)
plt.show()
```

**b)** Berechnen Sie die quadratische Abweichung $\mathrm{SE}\left(t\right)$ zwischen der numerischen und analytischen Lösung abhängig vom Zeitschritt:
   $$ 
   \mathrm{SE}\left(t\right) = \left(x_{\mathrm{numerisch}}\left(t\right) - x_\mathrm{analytisch}\left(t\right)\right) ^2 \, .
   $$
Plotten Sie die quadratische Abweichung gegen die Zeit. Benutzen Sie geeignete Beschriftungen! 

Beschreiben Sie kurz, was Sie beobachten. 

```python
#Numpy array und matplotlib importieren
import numpy as np
import matplotlib.pyplot as plt

#Definieren der Anfangswerte
x0 = 2
v0 = 1
dt = 0.005
a = -9.81

#Erstellen einer Liste für die quadratische Abweichung SE
listeSE = [] #leere Liste
for i in range (0, 81): #Schleife
    s = ((x0 + dt* i  * (v0 + 0.5 *dt *a*(i-1)))-(x0 + v0 * i*dt + 0.5 * a * (i*dt)**2))**2 #Befehl für die Schleife
    listeSE.append(s)

#Diagramm erstellen
fig, ax = plt.subplots()
ax.set_title("Quadratische Abweichung")  # Titel
#SE(t) plotten
ax.plot(listet, listeSE, label='SE(t)') 

ax.set_xlabel("Zeit $t$")
ax.set_ylabel("Quadratische Abweichung SE(t)")

plt.legend(loc=3)
plt.show()

#Beobachtung:
#Je größer der Zeitschritt ist, desto ungenauer ist die numerische Integration im Vergleich zur analytischen Lösung. 
#Der Fehler nimmt immer stärker zu.
```
