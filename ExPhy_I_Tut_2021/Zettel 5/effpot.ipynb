```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##constants
lam = 4
V0 = 10
L = 0.4
m = 0.5
r = np.linspace(0.11,1,100) #radius values

def eff(r):
    return -V0*np.exp(-lam**2*r**2)+L**2/(2*m*r**2)


plt.figure(dpi=200)
plt.grid()
#plt.xticks([], [])
#plt.yticks([], [])
plt.plot(r,eff(r),lw=1,label="$V_{eff}$")
plt.xlabel("$r$")
plt.ylabel("$V_{eff}$")
plt.hlines(np.min(eff(r)),r[0],r[-1],lw=1,color='g',label = "$E_0$")
plt.hlines(0.1,r[0],r[-1],lw=1,color='r',label = "$E_1$")
plt.hlines(np.max(eff(np.linspace(0.4,0.8,100))),r[0],r[-1],lw=1,color='c',label = "$E_2$")
plt.hlines(2,r[0],r[-1],lw=1,color='b',label = "$E_3$")
plt.legend()
plt.savefig("effpot.pdf")
```
