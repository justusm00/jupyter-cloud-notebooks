```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##constants
a = 0.5e-9 ##lattice constant
m = 150*1.66e-27 ##atomic mass
A1 = 1 ##amplitude
A2 = 1
```

```python
##atom amplitude
def u(q,wt,n):
    return A1 * np.exp(1j*(q*n*a-wt)) + A2 * np.exp(-1j*(q*n*a+wt)) 
```

```python
q = np.array([np.pi/a,np.pi/(2*a)]) ##wave vectors
wt = np.array([0,np.pi/2]) ##freq times time
n = np.arange(0,10,1) ##atom index
q_sy = [r'$\pi /a$',r'$\pi/2a$']

fig, ax = plt.subplots(2,2, figsize=(10,10))


ax[0,0].set_title("Re, $q=\pi/a$, $\omega t=0$")
ax[0,0].set_ylabel("$u(n)$")
ax[0,1].set_title("Im, $q=\pi/a$, $\omega t=0$")
ax[0,0].scatter(n,u(q[0],wt[0],n).real)
ax[0,1].scatter(n,u(q[0],wt[0],n).imag)

ax[1,0].set_title("Re, $q=\pi/a$, $\omega t=\pi/2$")
ax[1,0].set_ylabel("$u(n)$")
ax[1,0].set_xlabel("$n$")
ax[1,1].set_title("Im, $q=\pi/a$, $\omega t=\pi/2$")
ax[1,0].scatter(n,u(q[0],wt[1],n).real)
ax[1,1].scatter(n,u(q[0],wt[1],n).imag)
ax[1,1].set_xlabel("$n$")
```
