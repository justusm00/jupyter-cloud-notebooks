```python
import numpy as np
import matplotlib.pyplot as plt

#constants
Ef = 11.7 ##fermi energy in ev
Tf = 136e3 #fermi temperature in K
kB = 1.38e-23 #boltzmanns constant

def mu(T):
    return Ef*(1-np.pi**2/12*(T/Tf)**2)

Ts = np.array([0,300,10000])
print(mu(Ts))
```

```python
#Fermifunktion
def f(E,T):
    return 1/(np.exp((E-mu(T))/(kB*T*6.242e18))+1)

E = np.linspace(0.1,15,1000)

plt.figure(dpi=200)
plt.xlabel("$E$ in eV")
plt.ylabel("$f(E)$")
plt.grid()
for T in Ts:
    plt.plot(E,f(E,T),lw=1,label = "$T={}K$".format(T))
plt.legend()
```
