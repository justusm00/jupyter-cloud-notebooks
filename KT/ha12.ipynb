```python
import numpy as np
from numpy import exp
from numpy import cos
import matplotlib.pyplot as plt

t = np.linspace(0,15*1e-10,1000)
gs = 1/(9e-11) ##decay rate s
gl = 1/(5e-8) ##decay rate l
g = 0.5*(gs+gl)
dm = 5.3e9

def Ns(t):
    N = 1/4*(exp(-gs*t)+exp(-gl*t)+2*cos(dm*t)*exp(-g*t))
    Nbar = 1/4*(exp(-gs*t)+exp(-gl*t)-2*cos(dm*t)*exp(-g*t))
    Nks = 1/4*exp(-gs*t)
    Nkl = 1/4*exp(-gl*t)
    return N,Nbar,Nks,Nkl

    
N,Nbar,Nks,Nkl = Ns(t)
    
plt.figure(dpi=200)
plt.xlabel("$t$ in fs")
plt.ylabel("$N(t)$")
plt.grid()
plt.plot(t,N,lw=1,label="$K_0$")
plt.plot(t,Nbar,lw=1,label=r"$\overline{K_0}$")
plt.plot(t,Nks,lw=1,label="$K_S^0$")
plt.plot(t,Nkl,lw=1,label="$K_L^0$")
plt.legend()
```
