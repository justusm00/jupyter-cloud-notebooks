```python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8
a = (0.22*1e-15)**(-1)
R = np.sqrt(12)/a
hbar = 6.626e-34/(2*np.pi)
unit = 1e9*1.602e-19/c ##conversion factor to si units

q = np.linspace(0.1,5,int(1e4))*unit

def fac_sphere(Q):
    k = Q*R/hbar
    return 3/k**3*(np.sin(k)-k*np.cos(k))

def fac_prot(q):
    return 1/(1+(q/(a*hbar))**2)**2


fig, axs = plt.subplots(2, figsize= (7,7),dpi = 100)
plt.subplots_adjust(hspace = 0.5)

axs[0].grid()
axs[0].set_ylabel("$F(q)$")
axs[0].set_xlabel("$q$ in GeV/c")
axs[0].plot(q/unit,fac_sphere(q))
axs[0].set_title("Sphere")

axs[1].grid()
axs[1].set_ylabel("$F(q)$")
axs[1].set_xlabel("$q$ in GeV/c")
axs[1].plot(q/unit,fac_prot(q))
axs[1].set_title("Proton")


```
