```python
import numpy as np
import matplotlib.pyplot as plt

lam_a = 2.1e-3 #Zerfallskonstante Cd
lam_b = 8.955e-4 #Zerfallskonstante Ag
N0 = 1

def Ns(t):
    NA = N0 * np.exp(-lam_a*t)
    NB = lam_a / (lam_b-lam_a)*N0*(np.exp(-lam_a*t)-np.exp(-lam_b*t))
    NC = lam_a * lam_b / (lam_b-lam_a)*N0*(lam_b**(-1)*np.exp(-lam_b*t)-lam_a**(-1)*np.exp(-lam_a*t)) +N0
    return [NA,NB,NC]

t = np.linspace(0,5000,1000)

plt.figure(dpi=150)
plt.grid()
plt.title("Bruchteil der Kerne als Funktion der Zeit")
plt.xlabel("$t$ in s")
plt.ylabel("Bruchteil der Kerne")
plt.plot(t,Ns(t)[0], label = "Cd")
plt.plot(t,Ns(t)[1], label = "Ag")
plt.plot(t,Ns(t)[2], label = "Pd")
plt.legend(loc=1)
plt.savefig("zerfall.pdf")
```
