```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit



n_polym=20000
N=32
L=10
V=10*3*3
Nbar=n_polym/V
lam=0.00396
a=L*N/(lam*Nbar)

rates_eq=np.array([0.0005,0.001,0.002,0.003,0.004,0.005,0.006,0.007,0.008,0.009])
phiA=np.array([0.27160758, 0.18961236, 0.12205424 ,0.08755755 ,0.07120523 ,0.06244297,
 0.05647884 ,0.05038826, 0.04597011 ,0.04204938])



##analytical equation for phi0
def phi0(r,a):
    return 1/(2+a*r)




r = np.linspace(0,0.008,1000)

        

plt.figure(dpi=100)
plt.grid()
plt.title("Boundary densities as a function of the conversion rates")
plt.xlabel("Conversion rate")
plt.ylabel("Densities")
plt.plot(rates_eq,phiA,label="numerical")
plt.plot(r,phi0(r,a),label="analytical")

plt.legend(loc="upper right")
plt.show()
```
