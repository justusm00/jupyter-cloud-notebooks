```python
from scipy.stats import norm
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

plt.figure(dpi=400)

x = np.random.normal(-1,0.5,10000)
y = np.random.normal(1,0.5,10000)


plt.hist(x, bins = 100,alpha = 0.6,label = "$\mathcal{N}(x|-1,0.5)$",color = 'r')
plt.hist(y,bins = 100,alpha = 0.6, label = "$\mathcal{N}(y|1,0.5)$", color = 'b')
plt.hist(x*y,bins = 100,alpha = 0.5, label = "distribution of $xy$", color = 'g')
plt.hist(x+y,bins = 100,alpha = 0.5, label = "distribution of $x+y$", color = 'c')
plt.legend(loc = 'upper right')
```
