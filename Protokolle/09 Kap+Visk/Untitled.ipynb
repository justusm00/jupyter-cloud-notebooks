```python
import numpy as np

g=9.81

#systematischer Fehler
sigs = 0.005

##Massen
m1 = 0.050
m2 = 0.500
m3 = 5.000

##Radien in mm
##rot
rr = 0.5*np.array([22.46-21.53,22.42-21.535,22.39-21.512])
rrm = np.mean(rr)
rstd = np.std(rr)
sigr = np.sqrt(rstd**2+sigs**2)
##blau
rb = 0.5*np.array([22.66-21.42,24.41-22.66,-22.62+24.35])
rbm = np.mean(rb)
bstd = np.std(rb)
sigb = np.sqrt(bstd**2+sigs**2)
##gruen
rg = 0.5*np.array([22.835-20.95,22.91-20.81,22.875-20.82])
rgm = np.mean(rg)
gstd = np.std(rg)
sigg = np.sqrt(gstd**2+sigs**2)



##Berechnung der Dichten
##Wasser
rhow = 1000
Tw = m1 * 4 + m2 * 5 + m3 * 9
##Methanol
rhom = (7 * m3) / Tw * rhow
##Ethylenglukol
rhoe = (6 * m1 + 5 * m2 + 10 * m3) / Tw * rhow

##Fehlerfortplanzung Dichte
mges = m1 + m2 + m3
##Drehmom Wasser
tw = m1 * 4 + m2 * 5 + m3 * 9
##Systematischer Fehler
sigd = 0.5

##Fehler Dichte Methanol und Ethylenblablanol
sigme = sigd * rhow * np.sqrt((m3/tw)**2 + ((mges*m3*7)/(tw**2))**2)
siget = sigd * rhow * np.sqrt(((m1+m2+2*m3)/(tw))**2 + ((m1+m2+2*m3)*(6*m1+5*m2+10*m3)/(tw**2)**2))
print("Dichten mit Fehler:",rhom,rhoe,sigme,siget)

### Steighöhe

sigh = 0.5e-3
def height_mean(h):
    return np.mean(h)

def height_sigma(h):
    return np.sqrt(np.std(h)**2 + sigh**2)

##Wasser 
hw = np.array([height_mean([29e-3,25.5e-3,25e-3]), height_mean([7e-3,8e-3,7.5e-3]), height_mean([5.5e-3,6e-3,7e-3])])
sighw = np.array([height_sigma([29e-3,25.5e-3,25e-3]), height_sigma([7e-3,8e-3,7.5e-3]), height_sigma([5.5e-3,6e-3,7e-3])])

##Methanol
hm = 1/1000*np.array([height_mean([12,10,11]), height_mean([11,16,14]), height_mean([5,5,5.5])])
sighm = 1/1000*np.array([height_sigma([12,10,11]), height_sigma([11,16,14]), height_sigma([5,5,5.5])])

##Ethylen
het = 1/1000*np.array([height_mean([12.5,12.5,12]), height_mean([8,9,8.5]), height_mean([5,5,5])])
sighet = 1/1000*np.array([height_sigma([12.5,12.5,12]), height_sigma([8,9,8.5]), height_sigma([5,5,5])])

```

```python
##Berechnung der Oberflächenspannung

def tension(r,rho,h):
    return 0.5 * r * g* rho * h

##Radien signifikant gerundet
rr = 0.45e-3
sigrr = 0.02e-3
rb = 0.8e-3
sigrb = 0.2e-3
rg = 1e-3
sigrg = 0.05e-3
r = np.array([rr,rb,rg])
sigr = np.array([sigrr,sigrb,sigrg])

##Dichten signifikant gerundet
rhow = 1100
rhom = 730
sigrhom = 70
rhoet = 1100
sigrhoet = 110

tenw = np.array([tension(r,rhow,hw)])
tenm = np.array([tension(r,rhom,hm)])
tenet = np.array([tension(r,rhoet,het)])

def tension_sigma(r,rho,h,sigr,sigrho,sigh):
    return 0.5 * np.sqrt(sigr**2*(g*rho*h)**2 + sigrho**2*(g*r*h)**2 + sigh**2*(g*r*rho)**2)

sigtenw = tension_sigma(r,rhow,hw,sigr,0,sighw)
sigtenm = tension_sigma(r,rhom,hm,sigr,sigrhom,sighm)
sigtenet = tension_sigma(r,rhoet,het,sigr,sigrhoet,sighet)

print(tenw,sigtenw)
print(tenm,sigtenm)
print(tenet,sigtenet)
##gewichteter Mittelwert

num = 0
den = 0
for i in range(2):
    num += tenw[0][i]/sigtenw[i]**2
    den += 1/sigtenw[i]**2

tenw_ = num/den
sigtenw_ = np.sqrt(1/den)

num = 0
den = 0
for i in range(2):
    num += tenm[0][i]/sigtenm[i]**2
    den += 1/sigtenm[i]**2

tenm_ = num/den
sigtenm_ = np.sqrt(1/den)

num = 0
den = 0
for i in range(2):
    num += tenet[0][i]/sigtenet[i]**2
    den += 1/sigtenet[i]**2

tenet_ = num/den
sigtenet_ = np.sqrt(1/den)



print("gewichtet:",tenw_,sigtenw_,tenm_,sigtenm_,tenet_,sigtenet_)
```
