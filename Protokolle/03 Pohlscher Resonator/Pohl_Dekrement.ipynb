```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stat
from scipy import optimize
import math




###Berechnung des Dekrements
##0mm

sigs = 5 ##syst Fehler der Auslenkung
Max = np.array([110,90,80,65,50,40,])
Min = np.array([-105,-85,-70,-55,-40,-30])

d1 = [np.log(Max[i]/Max[i+1]) for i in range (len(Max)-1)]
d2 = [np.log(Min[i]/Min[i+1]) for i in range (len(Min)-1)]
print(d1)

##Fehler von d

sig1 = [math.sqrt(sigs**2*(1/Max[i])**2+sigs**2*(1/Max[i+1])**2) for i in range (len(Max)-1)]
sig2 = [math.sqrt(sigs**2*(1/Min[i])**2+sigs**2*(1/Min[i+1])**2) for i in range (len(Min)-1)]

##gewichteter Mittelwert
a = 0
b = 0
c = 0
d = 0
for i in range(len(Max)-1):
    a += d1[i]/(sig1[i])**2
    b += d2[i]/(sig2[i])**2
    c += 1/sig1[i]**2
    d += 1/sig2[i]**2

d1g = a/c
d2g = b/d
 
err1 = math.sqrt(1/c)
err2 = math.sqrt(1/d)


##absoluter Mittelwert
a = (d1g/err1**2)+(d2g/err2**2)
b = (1/err1**2)+(1/err2**2)
dg = a/b

errg = math.sqrt(1/b)

print(d1g,d2g)
print(err1,err2)
print(dg,errg)
```
