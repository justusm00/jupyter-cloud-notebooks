```python
import numpy as np
import matplotlib.pyplot as plt
n = 1.4560
l = 3e-3

I = np.array([5.36,6.16,7.15,8.17]) ##Spulenstrom in A
Ierr = 0.005 ##fehler des stroms
rerr = 7*1e-6 ##fehler der radiusquadrats


###Berechnung f端r I = 5.36 A
r11 = np.array([243888720,268693952]) / np.pi * 1e-12
r12 = np.array([214796416,243888720]) / np.pi * 1e-12

r21 = np.array([477204576,497545344]) / np.pi * 1e-12
r22 = np.array([449873120,477204576]) / np.pi * 1e-12

##funktion nimmt bereits die quadrate der radien
def diff(r11,r12,r21,r22):
    return 1/(2*n*l)*(r21/(r21-r11) - r22/(r22-r12))

##funktion um fehler zu returnen

def lamerr(r11,r12,r21,r22):
    a = (r11**2+r21**2) / (r21-r11)**2 + (r12**2+r22**2) / (r22-r12)**2
    return rerr/(2*n*l)*np.sqrt(a)

lam = diff(r11,r12,r21,r22)
err = lamerr(r11,r12,r21,r22)
print(lam,err)


```

```python
import scipy.optimize as opt
##Plots
B = np.array([0.32,0.36,0.41,0.46])
Berr = np.zeros(len(B))

lam = np.array([14.90,15.32,17.12,21.75])
lamerr = np.zeros(len(lam))

for i in range(len(Berr)):
    Berr[i] = 0.01
    lamerr[i] = 0.01

    
##regression
def f(x,m,b): 
	return  m*x+b
p0 = [0,0]
x_test = np.linspace(np.min(B),np.max(B),1000)
fit, err = opt.curve_fit(f, B,lam, p0, lamerr)
perr = np.sqrt(np.diag(err))


plt.figure(dpi = 400) #erstellen einer neuen Abbildung
#plt.title("Messung f端r verschiedene Anodenspannungen")
plt.xlabel("Magnetfeld $B$ in T") 
plt.ylabel("Differenz $\Delta(1/\lambda)$ in $\mathrm{m}^{-1}$") 
plt.grid()
plt.errorbar(B, lam, xerr = Berr ,yerr=lamerr, marker='x', ls='', capsize=2, capthick=0.7, lw=0.5, label="Messwerte")
plt.plot(x_test, f(x_test,fit[0],fit[1]), lw = 0.7,label="Fit",color = 'r') #Hiermit wird die Funktion dargestellt) 
plt.legend(loc = "upper left")
plt.savefig("normal.pdf")

print(fit,perr)
```
