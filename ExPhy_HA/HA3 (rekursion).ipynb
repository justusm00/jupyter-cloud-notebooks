```python
import numpy as np
a=-9.81
x_0=2

Δt=0.01

#setze Δt belieb klein, wobei 1/Δt eine ganze Zahl sein muss
##dauert länger, je kleiner Δt
def v(n):
    if n==0:
        return 1
    else:
        return(v(n-1)+a*Δt)
    
def x(n):
    if n==0:
        return 2
    else:
        return x(n-1)+v(n-1)*Δt
    


import numpy as np
import matplotlib.pyplot as plt

t_n = np.arange(0, 1,Δt)
n = np.linspace(0, 8, 9)
x_n = np.array([])
z=int(1/Δt)
print(z)
for i in range(0, z):
    xi = x(i)
    x_n = np.append(x_n, xi)
print(len(x_n))
print(len(t_n))
import matplotlib.pyplot as plt

def returnGraph(t):
    return 2+1*t+(a/2)*t**2
t = np.linspace(0, 5, 1000)

x = returnGraph(t)
fig, ax = plt.subplots()

ax.set_title("Bahnkurve zu x(t)")
ax.set_xlabel("t")
ax.set_ylabel("x(t)")
plt.axis([0, 0.8, 0, 2.5])
plt.grid(True)

ax.plot(t, returnGraph(t), label="x(t)", color="r")
ax.scatter(t_n, x_n, label='numerische Integration')

ax.legend()  
plt.show() 
```
