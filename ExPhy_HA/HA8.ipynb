## Rollen: Schiefe Ebene 

Wir betrachten einen Zylinder mit Radius $R$, Masse $m$ und Kreismittelpunkt $M$. Wir wollen die Bewegung des Zylinders entlang einer schiefen Ebene berechnen und visualisieren. Die schiefen Ebene hat einen Neigungswinkel $\theta = 12^{\circ}$. Der Startpunkt des Zylinders befinde sich $h_0=1$m über dem Boden.


<img src="./rollingPlane.png" alt="rollingPlane" width="300"/>

Die Höhe $h$ nach der auf der Ebene zurückgelegten Strecke $l$ beträgt
$$
h = h_0 - l \sin \theta
$$

und somit gilt für die potentielle Energie:
$$
E_{\mathrm{pot}} = m g (h_0 - l \sin \theta).
$$

Die kinetische Energie setzt sich zusammen aus der Translationsenergie $E_\mathrm{trans}=\frac{1}{2}mv^2 = \frac{1}{2}m\omega^2 R^2$ und der Rotationsenergie um die Hauptrotationsachse $E_\mathrm{rot}=\frac{1}{2}I\omega^2$, wobei $I$ das Trägheitsmoment des Zylinders bezüglich seiner Hauptrotationsachse ist
$$
E_\mathrm{kin}= E_\mathrm{trans} + E_\mathrm{rot} = \frac{1}{2}\left(I+mR^2\right)\omega^2.
$$

Die potentielle Energie wird beim Rollen in kinetische Energie umgesetzt. Es gilt also nach zurückgelegter Strecke $l$
$$
 m g l \sin \theta = \frac{1}{2}\left(I+mR^2\right)\omega^2.
 $$
 
**a)** Nehmen Sie an, es handle sich um einen Vollzylinder mit der Masse $m=1$ und einem Radius $R=0.1$.
Plotten Sie die Geschwindigkeit des Mittelpunktes $M$ in Abhängigkeit von der zurückgelegten Strecke $l$.

````python
import numpy as np
import math
import matplotlib.pyplot as plt
m=1
r=0.1
h0=1
theta=0.20943951023932
I=0.5*m*r**2
g=9.81
lmax=h0/np.sin(theta)
l=np.arange(0,lmax,0.001)
v= (((2*m*g*l*np.sin(theta))/(I+m*r**2))**(0.5))*r
plt.plot(l,v)
plt.grid()
````

**b)** Plotten Sie die $y$-Komponente des Punktes $A$ gegen die Zeit $t$ im mitbewegten Koordinatennsystem. Plotten Sie ebenfalls die Bahnkurve (also $x(t)$ gegen $y(t)$) des Mittelpunkts $M$ und die Bahnkurve des Punktes $A$ im ruhenden Bezugssystem.

````python
t = l/v
##y-Komponente von A
w=v/r
ya=-r*np.sin(w*t)
plt.figure()
plt.plot(t,ya)
plt.title("y-Komponente von A im bewegten System")
plt.xlabel("Zeit in s")
plt.ylabel("$y(t)$" )
plt.grid()

##Bahnkurve x/y
vt=(((2*m*g*v*t*np.sin(theta))/(I+m*r**2))**(0.5))*r
st=vt*t
x=np.cos(theta)*st
y=h0-np.sin(theta)*st
plt.figure()
plt.plot(x,y)
plt.title("Bahnkurve von M im ruhenden System")
plt.xlabel("x")
plt.ylabel("y" )
plt.grid()

##Bahnkurve von A
yat=y+ya
xat=st*np.cos(theta)+r*np.cos(w*t)
plt.figure()
plt.plot(xat,yat)
plt.title("Bahnkurve von A im ruhenden System")
plt.xlabel("x")
plt.ylabel("y" )
plt.grid()
````
