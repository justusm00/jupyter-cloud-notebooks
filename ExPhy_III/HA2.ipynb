```python
import numpy as np
import matplotlib.pyplot as plt


###aufgabe 3d
##Konstanten

k = 12
w = 5 
dk = 0.3
dw = 0.18


##Wellegleichungen
def psi1(x,t):
    return np.sin(x*(k-dk)-t*(w-dw))

def psi2(x,t):
    return np.sin(x*(k+dk)-t*(w+dw))

def psi(x,t):
    return 2*np.sin(k*x-w*t)*np.cos(dk*x-dw*t)


x = np.linspace(0,5,1000)


fig = plt.figure(dpi = 200)
ax = plt.axes()
plt.title("Zeitpunkt $t=0$")
plt.grid()
plt.ylabel('Auslenkung in m')
plt.xlabel("$x$")
ax.plot(x,psi1(x,0),linewidth= 1, label = "$\psi_1$")
ax.plot(x,psi2(x,0),linewidth= 1, label = "$\psi_2$")
ax.plot(x,psi(x,0),linewidth= 1, label = "$\psi$")
plt.legend(loc=1) 
plt.savefig("3d_t0.png")
```

```python
##Animation
from matplotlib import animation

t = np.linspace(-20,20,100)

fig = plt.figure(dpi = 200)
ax = plt.axes()
ax.set_aspect('equal')
line, = ax.plot([], [], lw=2)


# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
    ax.clear()
  

    ax.set_xlim(0,5)
    ax.set_ylim(-2,2)
    ax.plot(x,psi(x,t[i]),label = "$t=%f s$" %t[i],linewidth = 2)
    
    plt.title("wave")
    plt.grid()
    plt.ylabel('z')
    plt.xlabel('x')
    plt.legend(loc=1)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)

# save the animation as an mp4.  This requires ffmpeg or mencoder to be
# installed.  The extra_args ensure that the x264 codec is used, so that
# the video can be embedded in html5.  You may need to adjust this for
# your system: for more information, see
# http://matplotlib.sourceforge.net/api/animation_api.html
anim.save('wave_anim.mp4', fps=10, extra_args=['-vcodec', 'libx264'])

plt.show()
```

```python
##aufgabe 1c
a = 3
b = 12

def psi(x,t):
    return 5*np.exp(-a*x**2 - b * t**2 - 2*np.sqrt(a*b)*x*t)

x = np.linspace(-5,5,1000)
plt.figure(dpi = 200)

plt.grid()
plt.ylabel('Auslenkung in m')
plt.xlabel("$x$")
plt.plot(x,psi(x,0),linewidth= 1, label = "$t=0s$")
plt.plot(x,psi(x,2),linewidth= 1, label = "$t=2s$")
plt.legend(loc=1) 
plt.savefig("1c.png")
```
