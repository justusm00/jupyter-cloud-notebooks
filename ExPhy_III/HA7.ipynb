```python
import numpy as np
import matplotlib.pyplot as plt


f = 3
a = 1/(4*f)
def S(x,f,a):
    return f - x + np.sqrt((f-x)**2 + x / a)

x = np.linspace(-20,20,100)

plt.figure(dpi = 200)
plt.title("Reflexionsvermögen $R$")
plt.grid()
plt.ylabel("$R(n)$")
plt.xlabel("$n_I$")
plt.plot(x,S(x,f,a),linewidth= 1, label = "$n_R=1.5$")

plt.legend(loc = "upper right")
```
