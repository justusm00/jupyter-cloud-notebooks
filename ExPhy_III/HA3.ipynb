```python
import numpy as np
import matplotlib.pyplot as plt


c = 3e8
omega = 2.068*10**15
k = 4.449*10**12
lam = np.linspace(200,1600,1000)*10**(-9) ##wellenlnge in nm
w = 2*np.pi*c/lam


def eps_R(lam):
    w = 2*np.pi*c/lam
    return (1 - omega**2/(w**2+k**2))

def eps_I(lam):
    w = 2*np.pi*c/lam
    return 1 - omega**2*k/(w**3+k**2*w)



plt.figure(dpi = 200)
plt.title("Realteil")
plt.grid()
plt.ylabel('Re($\epsilon(\omega)$)')
plt.xlabel("Wellenlnge $\lambda$ in nm")
plt.plot(lam*1e9,eps_R(lam),linewidth= 1, label = "$\psi_1$")
plt.savefig("3c_Re.png")
```

```python
plt.figure(dpi = 200)
plt.title("Imaginrteil")
plt.grid()
plt.xlabel("Wellenlnge $\lambda$ in nm")
plt.ylabel('Im($\epsilon(\omega)$)')
plt.plot(lam*1e9,eps_I(lam),linewidth= 1)
plt.savefig("3c_Im.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

w0 = 1.88e16
lam  = np.linspace(380,780,1000)*1e-9
c = 3e8

def P(lam):
    w = 2*np.pi*c/lam
    return w**4/(w0**2+2*w**2)

plt.figure(dpi = 200)
plt.yticks([])
plt.title("Abgestrahlte Leistung als Funktion der Wellenlnge")
plt.grid()
plt.xlabel("Wellenlnge $\lambda$ in nm")
plt.ylabel("P")
plt.plot(lam*1e9,P(lam),linewidth= 1, label = "$\psi_1$")
plt.savefig("4d.png")
```
