```python
import numpy as np
import matplotlib.pyplot as plt

##3b

nI = np.linspace(1,25,1000)

def R(nR,nI):
    return ((1-nR)**2+nI**2)/((1+nR)**2+nI**2)

plt.figure(dpi = 200)
plt.title("Reflexionsvermgen $R$")
plt.grid()
plt.ylabel("$R(n)$")
plt.xlabel("$n_I$")
plt.plot(nI,R(1.5,nI),linewidth= 1, label = "$n_R=1.5$")
plt.plot(nI,R(0.1,nI),linewidth= 1, label = "$n_R=0.1$")
plt.legend(loc = "upper right")
plt.savefig("3b.png")
```

```python
##3c
lam = np.loadtxt("Ag.txt",usecols = 0,max_rows = 51)*1e3

#silber
nR_ag = np.loadtxt("Ag.txt",usecols = 1,max_rows = 51)
nI_ag = np.loadtxt("Ag.txt",usecols = 2,max_rows = 51)

##Gold
nR_au = np.loadtxt("Au.txt",usecols = 1,max_rows = 51)
nI_au = np.loadtxt("Au.txt",usecols = 2,max_rows = 51)

##Kupfer
nR_cu = np.loadtxt("Cu.txt",usecols = 1,max_rows = 51)
nI_cu = np.loadtxt("Cu.txt",usecols = 2,max_rows = 51)

plt.figure(dpi = 200)
plt.grid()
plt.ylabel("$R(\lambda)$")
plt.xlabel("Wellenlnge $\lambda$ in nm")
plt.errorbar(lam,R(nR_ag,nI_ag), linewidth= 1, label = "Ag",fmt = '+')
plt.errorbar(lam,R(nR_au,nI_au), linewidth= 1, label = "Au",fmt = '+')
plt.errorbar(lam,R(nR_cu,nI_cu), linewidth= 1, label = "Cu",fmt = '+')
plt.legend(loc = "upper right")
plt.savefig("3c.png")
```

```python
##3d
def T(nR,nI):
    return 1 - R(nR,nI)

plt.figure(dpi = 200)
plt.grid()
plt.ylabel("$T(\lambda)$")
plt.xlabel("Wellenlnge $\lambda$ in nm")
plt.errorbar(lam,T(nR_au,nI_au), linewidth= 1, label = "Au",fmt = '+')
plt.legend(loc = "upper right")
plt.savefig("3d.png")
```

```python
##4c

nP = 1.491
n = 1/nP
theta_max = np.arcsin(n)
theta = np.linspace(0,theta_max,1000)

##senkrecht
def Rs(theta):
    rs = -(np.cos(theta)-np.sqrt(n**2-np.sin(theta)**2))**2/(n**2-1)
    return rs**2

def Ts(theta):
    return 1- Rs(theta)

##parallel
def Rp(theta):
    rp = -(n**2*np.cos(theta)-np.sqrt(n**2-np.sin(theta)**2))/(n**2*np.cos(theta)+np.sqrt(n**2-np.sin(theta)**2))
    return rp**2
    
def Tp(theta):
    return 1 - Rp(theta)

plt.figure(dpi = 200)
plt.grid()
plt.xlabel("Winkel $\Theta_2$ in $\degree$")
plt.ylabel("Reflexionsgrad")
plt.plot(theta*180/np.pi,Rs(theta), linewidth= 1, label = "$R_\perp$")
plt.plot(theta*180/np.pi,Rp(theta), linewidth= 1, label = "$R_\parallel$")
plt.legend(loc = "upper right")
plt.savefig("4c_1.png")

plt.figure(dpi = 200)
plt.grid()
plt.xlabel("Winkel $\Theta_2$ in $\degree$")
plt.ylabel("Transmissionsgrad")
plt.plot(theta*180/np.pi,Ts(theta), linewidth= 1, label = "$T_\perp$")
plt.plot(theta*180/np.pi,Tp(theta), linewidth= 1, label = "$T_\parallel$")
plt.legend(loc = "upper right")
plt.savefig("4c_2.png")
```

```python
###4d 

alpha = 30/180*np.pi ##bogenma

plt.figure(dpi = 200)
plt.grid()
plt.xlabel("Winkel $\Theta_2$ in $\degree$")
plt.ylabel("Reflexionsgrad")
plt.plot(theta*180/np.pi,Rs(theta)*np.sin(alpha)**2, linewidth= 1, label = "$R_\perp$")
plt.plot(theta*180/np.pi,Rp(theta)*np.cos(alpha)**2, linewidth= 1, label = "$R_\parallel$")
plt.legend(loc = "upper right")
plt.savefig("4d.png")
```
