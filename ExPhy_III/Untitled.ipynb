```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert

lam = 500e-9
b = 13.5e-6
a = 75e-6
I0 = 1
theta = np.linspace(-np.pi/20,np.pi/20,1000)
k = np.sin(theta)*2*np.pi/lam

def sinc(a,b,k):
    return 16*np.sin(b*k/2)**2/(b**2*k**2)

def intense(a,b,k):
    return 16*np.cos(a*k/2)**2*np.sin(b*k/2)**2/(b**2*k**2)




plt.figure(dpi=200)
plt.title("sincc")
plt.xlabel("Wellenvektor $k$ in 1/m")
plt.ylabel("Normierte Intensitt")
plt.plot(k,intense(a,b,k),lw=1)
plt.plot(k,sinc(a,b,k),lw=1)
plt.savefig("2c.png")
plt.grid()
```
